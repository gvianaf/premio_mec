```python
def process_ideb_file(input_file, output_file, id_column, sheet_name=None):
    """
    Processa um arquivo IDEB, filtrando por 'Municipal' e salvando o resultado.

    Parâmetros:
    - input_file (str): Caminho do arquivo de entrada (Excel).
    - output_file (str): Caminho do arquivo de saída (CSV).
    - id_column (str): Nome da coluna principal ('ID_ESCOLA' ou 'CO_MUNICIPIO').
    """
    import pandas as pd

    # Ler o arquivo Excel com as colunas necessárias
    ideb_data = pd.read_excel(
        input_file,
        skiprows=9,
        sheet_name=sheet_name,
        usecols=[id_column, 'REDE', 'VL_OBSERVADO_2023']
    )
    # print(ideb_data.head(5))

    # Converter a coluna principal para numérico, tratar erros e remover NaN
    if id_column != 'SG_UF':
        ideb_data[id_column] = pd.to_numeric(ideb_data[id_column], errors='coerce')
        ideb_data = ideb_data.dropna(subset=[id_column])

    # Converter a coluna principal para int32
    if id_column != 'SG_UF':
        ideb_data[id_column] = ideb_data[id_column].astype('int32')

    # converte a coluna 'VL_OBSERVADO_2023' para float
    ideb_data['VL_OBSERVADO_2023'] = pd.to_numeric(ideb_data['VL_OBSERVADO_2023'], errors='coerce')
    ideb_data['VL_OBSERVADO_2023'] = ideb_data['VL_OBSERVADO_2023'].astype('float32')

    # Filtrar por 'Municipal' e remover a coluna 'REDE'
    # ideb_data = ideb_data[ideb_data['REDE'] == 'Municipal']
    # ideb_data = ideb_data.drop(columns=['REDE'])

    # Salvar o resultado em um arquivo CSV
    ideb_data.to_csv(output_file, sep=';', encoding='latin1', index=False)

# # Exemplo de uso com ID_ESCOLA:
# process_ideb_file(
#     input_file='ideb/divulgacao_anos_finais_escolas_2023.xlsx',
#     output_file='ideb/tratado/ideb_esc_af.csv',
#     id_column='ID_ESCOLA'
# )

# # Exemplo de uso com CO_MUNICIPIO:
# process_ideb_file(
#     input_file='ideb/divulgacao_anos_finais_municipios_2023.xlsx',
#     output_file='ideb/tratado/ideb_mun_af.csv',
#     id_column='CO_MUNICIPIO'
# )
```
